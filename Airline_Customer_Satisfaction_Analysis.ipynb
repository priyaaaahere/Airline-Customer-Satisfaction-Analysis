# Handling Missing Data and Data Cleaning

```python
#Checking for missing values in every column
print(df.isnull().sum())
```

```python
#To avoid dropping columns due to missing values, we'll replace null values with mean/median/mode
df.fillna({
    # Numerical columns
    'Customer ID': df['Customer ID'].median(),
    'Flight Distance': df['Flight Distance'].median(),
    'Departure Delay in Minutes': df['Departure Delay in Minutes'].median(),
    'Arrival Delay in Minutes': df['Arrival Delay in Minutes'].median(),
    'Age': df['Age'].median(),
    
    # Categorical columns->mode() 
    'Gender': df['Gender'].mode()[0],
    'Customer Type': df['Customer Type'].mode()[0],
    'Type of Travel': df['Type of Travel'].mode()[0],
    'Class': df['Class'].mode()[0],
    'Inflight wifi service': df['Inflight wifi service'].mode()[0],
    'Departure/Arrival time convenient': df['Departure/Arrival time convenient'].mode()[0],
    'Ease of Online booking': df['Ease of Online booking'].mode()[0],
    'Gate location': df['Gate location'].mode()[0],
    'Food and drink': df['Food and drink'].mode()[0],
    'Online boarding': df['Online boarding'].mode()[0],
    'Seat comfort': df['Seat comfort'].mode()[0],
    'Inflight entertainment': df['Inflight entertainment'].mode()[0],
    'On-board service': df['On-board service'].mode()[0],
    'Leg room service': df['Leg room service'].mode()[0],
    'Baggage handling': df['Baggage handling'].mode()[0],
    'Checkin service': df['Checkin service'].mode()[0],
    'Inflight service': df['Inflight service'].mode()[0],
    'Cleanliness': df['Cleanliness'].mode()[0],
    'satisfaction': df['satisfaction'].mode()[0],
}, inplace=True)
```

```python
print(df.isnull().sum())
```

# Categorical Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

#countplot is specifically used to visualize the distribution of categorical data. 
#each bar in a count plot represents the frequency of a unique category in the column.
categorical_columns = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']
#Count plots for categorical columns
for column in categorical_columns:
    plt.figure(figsize=(8,5))
    sns.countplot(x=column, data=df, palette='Set2')
    plt.title(f'Distribution of {column}')
    plt.show()
```

# Numerical Data Analysis

```python
#A histogram shows how values in a dataset are distributed across different intervals (bins)
numerical_columns = ['Flight Distance', 'Inflight wifi service', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
#Histogram for Numerical Data
plt.figure(figsize=(15, 10))
df[numerical_columns].hist(bins=20, figsize=(15, 10), layout=(2, 2), color='skyblue')
plt.show()
```

```python
#A box plot (also known as a box-and-whisker plot) is particularly useful for visualizing the distribution of numeric data, detecting outliers, 
#and understanding the spread of the data.

#Box plots 
for column in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[column], color='lightcoral')
    plt.title(f'Box Plot of {column}')
    plt.show()
```

```python
#A pair plot is particularly useful for visualizing relationships between pairs of numerical variables
#We used pairplot to:
#Visualizing Pairwise Relationships
#Detecting Linear and Non-Linear Relationships
#Identifying Correlation
#Exploring Data Distribution
#Detecting Clusters
#Checking for Outliers

#Pair plot to check relationship between numerical columns
sns.pairplot(df[numerical_columns])  #sns.pairplot(): This function from Seaborn is used to create a pair plot (also known as a scatterplot matrix)
plt.show()
```

```python
# Correlation matrix for numerical columns
correlation_matrix = df.corr() #calculates the correlation matrix for the numerical columns
#The correlation matrix is a table that shows the pairwise correlation coefficients between the numerical variables 

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')  #formatted with two decimal places
plt.title('Correlation Matrix')
plt.show()
```

```python
df['Total Delay'] = df['Departure Delay in Minutes'] + df['Arrival Delay in Minutes']

print(df[['Departure Delay in Minutes', 'Arrival Delay in Minutes', 'Total Delay']].head())

numerical_columns_with_total_delay = ['Flight Distance', 'Inflight wifi service', 'Departure Delay in Minutes', 'Arrival Delay in Minutes', 'Total Delay']

# Recalculate correlation matrix for the selected numerical columns
correlation_matrix = df[numerical_columns_with_total_delay].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix with Total Delay Feature')
plt.show()
```

# Customer Satisfaction Analysis

```python
# Satisfaction distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='satisfaction', data=df, palette='Set1')
plt.title('Customer Satisfaction Distribution')
plt.show()
```

```python
#The relationship between satisfaction and other categorical features
plt.figure(figsize=(15, 10))
categorical_columns = ['Gender', 'Customer Type', 'Type of Travel', 'Class']
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=column, hue='satisfaction', data=df, palette='Set2')
    plt.title(f'Satisfaction vs {column}')
    plt.legend(title='Satisfaction')
plt.tight_layout()
plt.show()
```

```python
#The relationship between satisfaction and numerical features
numerical_columns = ['Flight Distance', 'Total Delay', 'Inflight wifi service']
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='satisfaction', y=column, data=df, palette='Set3')
    plt.title(f'{column} by Satisfaction')
plt.tight_layout()
plt.show()
```

# Clustering Analysis(For Unsupervised Learning)

```python
#Importing the necessary libraries for clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

clustering_features = df[['Flight Distance', 'Inflight wifi service', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(clustering_features)

#K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

#Clusters using a pairplot
sns.pairplot(df, hue='Cluster', vars=clustering_features.columns, palette='Set2')
plt.show()
```

```python
# Count the number of data points in each cluster
print(df['Cluster'].value_counts())

# Analyze the relationship between clusters and satisfaction
plt.figure(figsize=(8, 5))
sns.countplot(x='Cluster', hue='satisfaction', data=df, palette='Set1')
plt.title('Customer Satisfaction across Clusters')
plt.legend(title='Satisfaction')
plt.show()
```
